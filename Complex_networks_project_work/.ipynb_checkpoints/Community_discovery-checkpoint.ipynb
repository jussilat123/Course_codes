```python
def community_strength(G, community_attr):
    # Get unique community names
    communities = set(attrs[community_attr] for node, attrs in G.nodes(data=True))
    print(communities)
    # Initialize a counter for number of internal and external links in different communities
    weak = {community: {'int': 0, 'ext': 0} for community in communities}
    # Assume all communities are strong
    strong = {community: True for community in communities}
    # Iterate over nodes
    for node in G.nodes():
        internal = 0
        external = 0
        # Get node community
        node_community = G.nodes[node][community_attr]
        # Iterate over nodes neighbors
        for neighbor in G[node]:
            # If neighbor is in the same community increase internal otherwise external link counts
            if node_community == G.nodes[neighbor][community_attr]:
                internal +=1
            else:
                external +=1
        weak[G.nodes[node][community_attr]]['int'] = internal
        weak[G.nodes[node][community_attr]]['ext'] = external
        if internal <= external:
            strong[node_community] = False
            print(f'Strong community test fails for node {node} in community {node_community}')
    weak = {community: weak[community]['int'] > weak[community]['ext'] for community in weak}
    return {'weak': weak, 'strong': strong}


#seems that Emails network doesn't have strong communities but it does have several weak communities.
weak_strong_communites = community_strength(Emails, 'department')
print(weak_strong_communites['weak'])

#collect  weak communities for list
Email_weak_communities = []
for key in weak_strong_communites['weak']:
    if(weak_strong_communites['weak'][key]):
        Email_weak_communities.append(key)
        
print(Email_weak_communities)
#19 out of 42 departments are weak communities. So Email network does have some community properties
```

```python
#DEMON
#g = nx.karate_club_graph()
dm = d.Demon(graph=Emails, epsilon=0.65, min_community_size=3)
coms = dm.execute()
```

```python
#community sizes
sizes = np.shape(coms)
print(sizes)
for Demon_k in range(sizes[0]):
    print("Demon_k: ",Demon_k," ",np.shape(list(coms[Demon_k])))
```

```python
def compute_stats(community_values,attributes,name_of_algorithm):
    
    nodes_in_strong_communities = {}
    
    found_community_as_real_apartment = []
    
    #count department values
    counter_attr = Counter(attributes[:,1])
    counter_attr = sorted(counter_attr.items())
    counter_attr_list = list(counter_attr)
    
    #convert lists into dictionaries
    attributes = dict(attributes)
    counter_attr_list = dict(counter_attr_list)
    
    #iterate all communities
    sizes = np.shape(community_values)
    for community in range(sizes[0]):
        
        #count departments in given community
        departments_in_community = []
        for node_community in list(community_values[community]):
            departments_in_community.append(attributes[int(node_community)])
        counter = Counter(departments_in_community)
        counter_list = list(counter.most_common())
        
        #compute some stats and print them
        found_community_as_real_apartment.append(counter_list[0][0])
        department_percent = counter_list[0][1]/len(departments_in_community)*100
        most_common_percent_to_all_depts = counter_list[0][1]/counter_attr_list[counter_list[0][0]]*100
        print(name_of_algorithm, " community ",community)
        print("The most common department value for this ", name_of_algorithm ," was ",counter_list[0][0])
        print("The ", name_of_algorithm ," contais ",counter_list[0][1], " units of the most of common department of this ",name_of_algorithm)
        print("The most common department value was {:.1f}% of this community".format(department_percent))
        print("This ", name_of_algorithm ," had department value ",counter_list[0][0], " {:.1f}% of all this department".format(most_common_percent_to_all_depts))
        print(" ")
        
        #if 50% or more of this community is from one department, save these nodes which are most common
        if(department_percent >= 50):
            for node_community in list(community_values[community]):
                if(attributes[int(node_community)] == counter_list[0][0]):
                    dept_num = attributes[int(node_community)]
                    if(dept_num not in nodes_in_strong_communities):
                        nodes_in_strong_communities[dept_num] = []
                    
                    if(node_community not in nodes_in_strong_communities[dept_num]):
                        nodes_in_strong_communities[dept_num].append(node_community)
                    #if (node_community in nodes_in_strong_communities):
                    #    nodes_in_strong_communities[node_community] += 1
                    #else:
                    #    nodes_in_strong_communities[node_community] = 1
        
        
    return found_community_as_real_apartment,nodes_in_strong_communities
```

```python
attributes  = np.loadtxt('email-Eu-core-department-labels.txt',dtype = int)
demon_community_departments,dictionary_demon = compute_stats(coms,connected_attributes,"Demon community")
#print(demon_community_departments)
Counter(demon_community_departments).most_common() #shows how many times this department was most common community in demon algorithm
#seems that many communities overlap
print(dictionary_demon)
for key in dictionary_demon:
    print("Dept: ",key," size: ",len(dictionary_demon[key]))
```

```python
#infomap
from infomap import Infomap

#import edge_list into numpy array
edge_list = []
f = open('Emails_connected.txt','r')
for line in f:
    splitted = line.split(" ")
    edge_list.append((int(splitted[0]),int(splitted[1])))
f.close()

edge_list_np = np.array(edge_list)

im = Infomap()

#add links to infomap
for link_pair in edge_list_np:
    im.add_link(link_pair[0],link_pair[1])
    
# Run the Infomap search algorithm to find optimal modules
im.run()

infomap_modules = []
module_list = []
previous_module = 1
print(f"Found {im.num_top_modules} modules with codelength: {im.codelength}")

#save node_ids into list infomap_modules
nodes = 0
for node in im.tree:
    if node.is_leaf:
        nodes += 1
        if(node.module_id != previous_module):
            previous_module = node.module_id
            infomap_modules.append(module_list)
            module_list = []
        else:
            module_list.append(node.node_id)
print("nodes: ",nodes)
```

```python
departments,dictionary_infomap = compute_stats(infomap_modules,attributes,"Infomap")
print(dictionary_infomap)
for key in dictionary_infomap:
    print("Dept: ",key," size: ",len(dictionary_infomap[key]))
```
